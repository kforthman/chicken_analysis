```python
import pandas as pd
import numpy as np
import scipy as sp
import plotly.plotly as py
import plotly.figure_factory as ff
import plotly.graph_objs as go
import json
```

#### Simple Plot

First, read in the data.

```python
df = pd.read_csv("../Output/Simulated_nl_Analysis/simAvg.csv")
df = df.drop(columns = ['subID'])
df.head()
```

```python
df_subset = df.loc[
            (df['N'] == 50) & 
            (df['sigma'] == 160) & 
            (df['H_true'] == 0.05) &
            (df['noise'] == 0.01) &
            (df['lapse'] == 0)
]
df_subset
```

```python
trace0 = go.Scatter(
    x = df_subset.H_subj,
    y = df_subset.H_subj_EST_avg,
    error_y = dict(
        type = 'data',
        array = df_subset.H_subj_EST_stder,
        visible = True
    ),
    mode = 'markers'
)
trace1 = go.Scatter(
    x = [0, 1],
    y = [0, 1],
    mode = 'lines'
)
data = [trace0, trace1]

py.plot(data, filename = 'basic-line', auto_open=True)
```

#### Grids

```python
from plotly.grid_objs import Grid, Column
import time
```

```python
sigma_range = np.array([110, 160])
Htrue_range = np.array([0.05, 0.95])
Hsubj_range = np.arange(0.05, 0.95, 0.05)
lapse_range = np.arange(0, 0.25, 0.05)
noise_range = np.arange(0, 6, 0.5)
noise_range[0] = 0.01

column_1 = Column(sigma_range, 'sigma')
column_2 = Column(Htrue_range, 'Htrue')
column_3 = Column(Hsubj_range, 'Hsubj')

grid = Grid([column_1, column_2, column_3])
py.grid_ops.upload(grid, 'ping_pong_grid'+str(time.time()), auto_open=False)
```

```python
data = [dict(
        visible = False,
        line=dict(color='#00CED1', width=6),
        name = 'ùúà = '+str(thisSigma),
        x = df.loc[
            (df['sigma'] == thisSigma) & 
            (df['H_true'] == 0.95) &
            (df['noise'] == 0.01) &
            (df['lapse'] == 0),
            ['H_subj']].values,
        y = df.loc[
            (df['sigma'] == thisSigma) & 
            (df['H_true'] == 0.95) &
            (df['noise'] == 0.01) &
            (df['lapse'] == 0),
            ['H_subj_EST']].values) for thisSigma in np.array([110, 160])] for thisH_true in np.array([0.05, 0.95])
```

```python
data[1]['visible'] = True
```

```python


sliders = []

sigma_range = np.array([110, 160])
Htrue_range = np.array([0.05, 0.95])
Hsubj_range = np.arange(0.05, 0.95, 0.05)
lapse_range = np.arange(0, 0.25, 0.05)
noise_range = np.arange(0, 6, 0.5)
noise_range[0] = 0.01

sliders.append(
    dict(
        active = sigma_range[0]
        currentvalue = {"prefix": "Sigma: "}
        steps = [{
            
        }]
    )
)


for i in range(len(data)):
    step = dict(
        method = 'restyle',
        args = ['visible', [False] * len(data)],
    )
    step['args'][1][i] = True # Toggle i'th trace to "visible"
    steps.append(step)

sliders = [dict(
    active = 10,
    currentvalue = {"prefix": "Frequency: "},
    pad = {"t": 50},
    steps = steps
)]

layout = dict(sliders=sliders)
fig = dict(data=data, layout=layout)

py.iplot(fig, filename='Sine Wave Slider')
```
